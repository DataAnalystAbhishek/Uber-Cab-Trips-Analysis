Uber has been a major source of travel for people living in urban areas. Some people don’t have their vehicles while some don’t drive their vehicles intentionally because of their busy schedule. So different kinds of people are using services of Uber and other taxi services.

By analyzing Uber trips, we can draw many patterns like which day has the highest and the lowest trips or the busiest hour for Uber and many other patterns. The dataset I’m using here is based on Uber trips from New York, a city with a very complex transportation system with a large residential community.

```python
# I will start this task of Uber cab Trips analysis by importing the necessary Python libraries and the dataset:

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
data = pd.read_csv("C:/Users/ABHISHEK TAILOR/Desktop/uber_trips_2014.csv")
data.head()
```

let’s prepare the data that I am using here to analyze the Uber trips according to days and hours:

```python
import datetime
data["Date/Time"] = data["Date/Time"].map(pd.to_datetime)
```

```python
data["Day"] = data["Date/Time"].apply(lambda x: x.day)
data["Weekday"] = data["Date/Time"].apply(lambda x: x.weekday())
data["Hour"] = data["Date/Time"].apply(lambda x: x.hour)
print(data.head())
```

So I have prepared this data according to the days and hours, as I am using the Uber cab trips for the September month so let’s have a look at each day to see on which day the Uber cab trips were highest:

```python
width=8
hight=6
sns.set(rc={'figure.figsize':(width, hight)})
sns.distplot(data["Day"],color='red')
```

By looking at the daily trips we can say that the Uber trips are rising on the working days and decreases on the weekends. Now let’s analyze the Uber trips according to the hours:

```python
sns.distplot(data["Hour"],color='green')
```

According to the hourly data, the Uber cab trips decreases after midnight and then start increasing after 5 am and the trips keep rising till 6 pm such that 6 pm is the busiest hour for Uber then the trips start decreasing. Now let’s analyze the Uber cab trips according to the weekdays:

```python
sns.distplot(data["Weekday"],color='blue')
```

In the above figure 0 indicates Sunday, on Sundays the Uber cab trips and more than Saturdays so we can say people also use Uber for outings rather than for just going to work. On Saturdays, the Uber trips are the lowest and on Mondays, they are the highest. Now let’s have a look at the correlation of hours and weekdays on the Uber cab trips:

```python
# Correlation of Weekday and Hour
df = data.groupby(["Weekday", "Hour"]).apply(lambda x: len(x))
df = df.unstack()
sns.heatmap(df, annot=False)
```

As we are having the data about longitude and latitude so we can also plot the density of Uber trips according to the regions of the New Your city:

```python
data.plot(kind='scatter', x='Lon', y='Lat', alpha=0.4, s=data['Day'], label='Uber Trips',
figsize=(8, 6), cmap=plt.get_cmap('jet'))
plt.title("Uber Cab Trips Analysis")
plt.legend()
plt.show()
```
